```python
#Let us Import the Book We would be taking under Consideration
Book = open(r"C:\Users\LENOVO\Downloads\The-Monk-Who-Sold-His-Ferrari.txt",encoding = 'utf-8')
```

```python
#Reading the content in the book and spliting it over list of lines
InitialReading = Book.read().splitlines()
```

```python
#removing the empty lines
InitialReading = [i for i in InitialReading if i !='']
#joining the non-empty lines into one single string
Text =""
Text = Text.join(InitialReading)
```

```python
#Printing the text after removing the empty spaces 
print(Text)
```

```python
#Data Preprocessing With Removal of unnecessary Characters, punctuations
Tofilter = '''!()-[]{};:'"\,<>./‘’?“”@#$%^&*_~'''
FilteredText = ""
for j in Text:
    if j not in Tofilter:
        FilteredText = FilteredText + j
```

```python
#Converting the text in Lowercase
FilteredText = FilteredText.lower()
print(FilteredText)
```

```python
#Let us Install Important Libraries and Resources needed for further advancement
import matplotlib.pyplot as plt
import pandas as pd
import nltk
import re
import string
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
#Now it is time to tokenize the Pre Processed text 
Tokens = word_tokenize(FilteredText)
print(Tokens)
```

```python
print(len(Tokens))
```

```python
#Let us check the frequency distribution of Tokens now
Frequency = nltk.FreqDist(Tokens)
```

```python
for token, frequency in Frequency.items():
    print(f"{token}: {frequency} times")
```

```python
print(Frequency.most_common(15))
Unique_Tokens = list(Frequency)
```

```python
#No of unique words/tokens in the text
print(len(Unique_Tokens))
```

```python
#Plotting tokens vs its frequency
import seaborn as sb
sb.set(style = 'whitegrid')
dataf = pd.DataFrame(Tokens)

sb.countplot(x=dataf[0],order=dataf[0].value_counts().iloc[:25].index)
plt.xticks(rotation=90)
plt.xlabel('words')
plt.show()
```

```python
#Construction of WordCloud Without Removing Stopwords
Stopwordswordcloud = WordCloud(width = 800, height = 600, 
                background_color ='black', 
                min_font_size = 10,stopwords = {},colormap='winter').generate(FilteredText)
plt.figure(figsize = (10,8), facecolor = None) 
plt.imshow(Stopwordswordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show() 
```

```python
#For better analysis it is expected to remove stopwords
stop_words = set(stopwords.words('english'))
tokens = word_tokenize(FilteredText)
final_tokens = [i for i in tokens if not i in stop_words] 
FinalText = "  "
FinalText = FinalText.join(final_tokens)
```

```python
# Word cloud after Stopwords removal
wc_withoutStopwords = WordCloud(width = 800, height = 600, 
                background_color ='black', 
                min_font_size = 10,stopwords = {},colormap='winter').generate(FinalText) 

plt.figure(figsize = (12,8), facecolor = None) 
plt.imshow(wc_withoutStopwords) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show() 
```

```python
#Plotting Words/tokens after removal of stopwords VS Frequency
Tokens_new = word_tokenize(FinalText)
FrequencyNew = nltk.FreqDist(Tokens_new)

new_dataframe=pd.DataFrame(Tokens_new)
sb.countplot(x=new_dataframe[0], order=new_dataframe[0].value_counts().iloc[:35].index)
plt.xticks(rotation=90)
plt.xlabel('Words')
plt.show()
```

```python
#Let us do PosTagging 
PosTagging = nltk.pos_tag(Tokens_new)
from collections import Counter
counts = Counter( tag for word,  tag in PosTagging)
print(counts)
```

```python
#Plotting Frequency Distribution of PoS Tags

pos_tags_freq = nltk.FreqDist(counts)
pos_tags_freq = {k: v for k, v in sorted(pos_tags_freq.items(), key=lambda item: item[1],reverse=True)}
x = list(pos_tags_freq.keys())[:40]
y = list(pos_tags_freq.values())[:40]
plt.figure(figsize=(12,5))
plt.plot(x,y,c='r',lw=4,ls='-.')
plt.grid()
plt.xticks(rotation=90)
plt.title('Frequency Distribution of PoS Tags',size=15)
plt.xlabel('Words',size=14)
plt.ylabel('Frequency',size=14)
plt.show()
```

```python
#Now let us Open the largest chapter of the book chapter 9(The Ancient Art of Self Leadership)
Book1 = open(r"C:\Users\LENOVO\Downloads\LargestChapter.txt",encoding = 'utf-8')
```

```python
InitialReading1 = Book.read().splitlines()
InitialReading1 = [i for i in InitialReading if i !='']
#joining the non-empty lines into one single string
Text1 =""
Text1 = Text1.join(InitialReading1)
```

```python
#Data Preprocessing With Removal of unnecessary Characters, punctuations
Tofilter = '''!()-[]{};:'"\,<>./‘’?“”@#$%^&*_~'''
FilteredText1 = ""
for j in Text1:
    if j not in Tofilter:
        FilteredText1 = FilteredText1 + j
```

```python
FilteredText1 = FilteredText1.lower()
```

```python
Tokens1 = word_tokenize(FilteredText1)
cleaned_tokens = [token for token in Tokens1 if token != "<<UNK>>"]
```

```python
from nltk import bigrams
bigram_list = list(bigrams(cleaned_tokens))
bigram_list
```

```python
print(len(bigram_list))
```

```python
bigram_freq1 = nltk.FreqDist(bigram_list)
bigram_probabilities = {}
for bigram in bigram_freq1:
    first_word = bigram[0]
    probability = bigram_freq1[bigram]/Tokens1.count(first_word)
    bigram_probabilities[bigram] = probability
bigram_probabilities = sorted(bigram_probabilities.items(),key = lambda x: x[1], reverse = True)
#Printing the values
for bigram, probability in bigram_probabilities[:50]:
    print(f"{bigram[0]} {bigram[1]}: {probability}")
```
